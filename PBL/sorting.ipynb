```python
def binary_search(sequence, item):
    begin_idx = 0
    end_idx = len(sequence) - 1

    while begin_idx <= end_idx:
        midpoint = begin_idx + (end_idx - begin_idx) // 2
        midpoint_val = sequence[midpoint]
        if midpoint_val == item:
            return midpoint
        
        elif item < midpoint_val:
            end_idx = midpoint - 1
        
        else:
            begin_idx = midpoint + 1
    
    return None

sequence_a = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
item_a = 23
hasil = binary_search(sequence_a, item_a)

if hasil is not None:
    print(f"Angka {item_a} ditemukan pada indeks ke-{hasil}")
else:
    print(f"Angka {item_a} tidak ditemukan")
```

```python
def selection_sort(arr):
    for i in range(0, len(arr) - 1):
        cur_min_idx = i
        for j in range (i + 1, len(arr)):
            if arr[j] < arr[cur_min_idx]:
                cur_min_idx = j
        arr[i], arr[cur_min_idx] = arr[cur_min_idx], arr[i]
    return arr
#arr = [2, 6, 5, 1, 3, 4]
arr = [4, 3, 2, 1]
selection_sort(arr)
print(f"Hasilnya adalah: {arr}")
```

```python
def selectionSort(nums):
    n =  len(nums)

    for i in range (n):
        min_index = i
        for j in range (i + 1, n):
            if nums[j] < nums[min_index]:
                min_index = j
        nums[i], nums[min_index] = nums[min_index], nums[i]
    return nums

nums = [11, 34, 25, 12, 22, 64, 9]
selectionSort(nums)
print(f"Hasilnya adalah: {nums}")
```

```python
nums = [1, 6, 3, 4, 5, 2, 2]
def findDuplicate(nums):
    slow = nums[0]
    fast = nums[0]

    #fase 1
    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break
    #fase 2
    slow = nums[0]
    while slow != fast:
        slow = nums[slow]
        fast = nums[fast]

    return slow

hasil = findDuplicate(nums)
print(f"Duplikatnya adalah: {hasil}")
```
